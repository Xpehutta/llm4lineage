```python
import pandas as pd
from pathlib import Path

from Classes.regexp_extractor import RegexSQLExtractor
from Classes.model_classes import SQLLineageResult
```

```python
# Get the current working directory
current_dir = Path.cwd()
# Construct the path relative to current directory
file_path = current_dir / 'data' / 'SQL.txt'

# Read file with example
with open(file_path, "r", encoding="utf-8") as f:
    SQL = f.read()
```

```python
print(f"\n   Testing Regexp extraction with query ...")
re_extractor = RegexSQLExtractor()
target = re_extractor.extract(SQL)

print(f"✓ Successfully extracted lineage")
print(f"   Target: {target.get('target', 'N/A')}")
print(f"   Sources: {len(target.get('sources', []))}")

# Create SQLLineageResult object
lineage_result = SQLLineageResult(
    target=target.get('target', ''),
    sources=target.get('sources', [])
)

print(f"   As JSON: {lineage_result.to_json()}")
```
